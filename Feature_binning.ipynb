```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Churn_Modelling.csv')
data.head()
```

```python
data.drop(columns=['CustomerId', 'RowNumber', 'Surname'], axis=1, inplace=True)
```

FEATURE BINNING BASED ON AGE

```python
labels = ['0-20', '21-40', '41-60', 'Above 61']
bins = [0, 20, 40, 60, 100 ]

data['Age_bins'] = pd.cut(data.Age, bins, labels = labels, include_lowest=True)
# here we are adding respective bin to every row
```

```python
data.head()
```

```python
data['Age_bins'].value_counts()
```

```python
plt.bar(labels, data.Age_bins.value_counts())

# giving the title
plt.title('Age Count')

# giving X & Y labels
plt.xlabel('Age Bins')
plt.ylabel('Age Count')

# visualizing the plot
plt.show
```
