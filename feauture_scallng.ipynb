```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Churn_Modelling_With_Nulls.csv')
df.info()
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df.describe()
```

```python
newdf = pd.DataFrame(df,columns=['CreditScore','Tenure','Balance'])
newdf.info()
newdf.head()
```

```python
newdf['Balance'] = newdf['Balance'].fillna(method='bfill')
newdf.info()
```

```python
newdf['CreditScore']=newdf['CreditScore'].fillna(newdf['CreditScore'].mean())
newdf.info()
```

```python
# MIN MAX scaling method

scaler = MinMaxScaler()
minmaxdf = scaler.fit_transform(newdf)
print(minmaxdf)
```

```python
# standard scalar method

scaler = StandardScaler()
scaledf = scaler.fit_transform(newdf)
print(scaledf)
```
